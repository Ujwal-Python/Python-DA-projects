## Irrelavent columns (correlation)

### Removing Irrelevant Columns

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

### Renaming the column

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

### Dropping the duplicate row

```python
df.shape
```

```python
duplicate_row_df = df[df.duplicated()]
print("Number of duplicate rows: ", duplicate_row_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head()
```

```python
df.count()
```

### Dropping the missing or null values

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.count()
```

```python
df.isnull().sum()
```

### Checking for outliers

```python
sns.boxplot(x = df['Price'])
```

```python
sns.boxplot(x =df['HP'])
```

```python
sns.boxplot(x= df['Cylinders'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~ ((df < (Q1 - 1.5*IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis = 1) ]
df.shape
```

### Histogram
Histogram refers to the frequency of occurrence of variables in an interval. There are mainly 10 different types of car manufacturing companies, but it is often important to know who has the most number of cars.

```python
df["Make"].value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

### Heatmap
For checking the dependent variable / Correlation between variable.

```python
plt.figure(figsize= (10,5))
c = df.corr()
sns.heatmap(c,annot = True)
c
```

### Scatter Plot
We are using this for finding the correlation between the variables.

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```
